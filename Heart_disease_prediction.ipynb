#Heart Disease Prediction

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

##Loading and PreProcessing Data

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Heart Disease Prediction/data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['StressLevel'].replace({'Low': 1, 'Medium': 2, 'High': 3}, inplace = True)
```

```python
df['StressLevel'].value_counts()
```

```python
df['HeartDisease'].value_counts()
```

```python
df.drop_duplicates(inplace = True)
df.shape
```

```python
df.corr()
```

##Splitting Data into training and Cross-validation set

```python
x = df.drop(columns = 'HeartDisease', axis = 1).to_numpy()
y = df['HeartDisease']
y = y.to_numpy()
print(x.shape)
print(y.shape)
```

```python
x_train, x_cv, y_train, y_cv = train_test_split(x, y, test_size = 0.2,stratify = y, random_state = 1)
print(x_train.shape)
print(y_train.shape)
print(x_cv.shape)
print(y_cv.shape)
```

##Logistic Regression

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_cv_scaled = scaler.transform(x_cv)
```

```python
lr_model = LogisticRegression()
lr_model.fit(x_train_scaled, y_train)
```

```python
yhat = lr_model.predict(x_train_scaled)
accuracy_score(yhat, y_train)
```

```python
yhat =lr_model.predict(x_cv_scaled)
accuracy_score(yhat, y_cv)
```
